```python
# James Dietle code to seperate drivers for statefarm
# https://www.kaggle.com/c/state-farm-distracted-driver-detection

import pandas as pd
import random
import os

# Where the data is stored
path = '/home/jd/data/statefarm/'

# read the CSV file into a pandas DataFrame
driversData = pd.read_csv(path + 'driver_imgs_list.csv')

# group the frame by the subject in the image
myset= set(list(driversData.subject))

# Number of drivers to be moved;
drivers_moved= 3
valid_set=random.sample(set(myset), drivers_moved)
print ("Drivers that will be moved " + str (valid_set))

# List to filenames associated with drivers
filenames = list()
for i in list(valid_set):
    # Of the drivers selected, writes out their images, and folder
    temp = list((driversData.loc[driversData['subject'] == i].classname) + driversData.loc[driversData['subject'] == i].img)
    filenames= filenames + temp 
```

```python
# Source of the training data to be moved
source = '/home/jd/data/statefarm/train/'
# Destination for validation set
dest = '/home/jd/data/statefarm/valid/'

# Makes the move command
for i in filenames:
    folder= i[0:2]
    file = i[2:]
    print('mv '+ source + folder + "/"+file +" "+dest+folder+"/"+file)
```
